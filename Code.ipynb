# Compute data (dis-)similarity

### Nominal Features

Function to find similarity between nominal features

```python
# create the function for nominal features similarity

import itertools

def nominal_sim(nom1):
    similarities = []
    for x,y in itertools.combinations(nom1, 2):
        if x == y:
            similarities.append(1)
        else: 
            similarities.append(0)
    return similarities 
```

```python
# find the dissimilarity vectors per nominal feature

disim1 = [1 - x for x in nominal_sim(groceries_data.Sex)]
disim2 = [1 - x for x in nominal_sim(groceries_data.Marital_Status)]
disim3 = [1 - x for x in nominal_sim(groceries_data.Occupation)]
```

```python
# transform the vectors into square matrices

from scipy.spatial.distance import *

disim_matrix_sex = squareform(disim1)
disim_matrix_marital = squareform(disim2)
disim_matrix_occupation = squareform(disim3)
```

```python
# create the column and row names for presentation purposes

cust_ids = ['Cust' + str(x2) for x2 in groceries_data.Customer_ID]
cust_ids[0:3]
```

```python
# transform the square matrices into lower triangle matrices 

import numpy as np

df1_sex = pd.DataFrame(np.tril(disim_matrix_sex), columns=cust_ids, index=cust_ids) 
df2_marital = pd.DataFrame(np.tril(disim_matrix_marital), columns=cust_ids, index=cust_ids) 
df3_occupation = pd.DataFrame(np.tril(disim_matrix_occupation), columns=cust_ids, index=cust_ids) 
```

```python
# print the shape and a part of the matrix to check the rightness of the code

print(df1_sex.shape)
df1_sex.iloc[0:5,0:5]
```

### Numerical features 

Function to find similarity between numerical features

```python
# create the function for numerical features similarity

def numerical_sim(num1):
    max_value = max(num1.unique())
    min_value = min(num1.unique())
    disimilarities = []
    for x,y in itertools.combinations(num1, 2):
        disimilarities.append(abs(x-y)/(max_value - min_value))
    return disimilarities 
```

```python
# find the dissimilarity vectors per numerical feature

disim1b = numerical_sim(groceries_data.Persons_in_Household)
disim2b = numerical_sim(groceries_data.Age)
disim3b = numerical_sim(groceries_data.Income)
```

```python
# transform the vectors into square matrices 

from scipy.spatial.distance import *

disim_matrix_persons = squareform(disim1b)
disim_matrix_age = squareform(disim2b)
disim_matrix_income = squareform(disim3b)
```

```python
# transform the square matrices into lower triangle matrices

import numpy as np

df1_persons = pd.DataFrame(np.tril(disim_matrix_persons), columns=cust_ids, index=cust_ids) 
df2_age = pd.DataFrame(np.tril(disim_matrix_age), columns=cust_ids, index=cust_ids) 
df3_income = pd.DataFrame(np.tril(disim_matrix_income), columns=cust_ids, index=cust_ids) 
```

```python
# print the shape and a part of the matrix to check the rightness of the code

print(df2_age.shape)
df3_income.iloc[0:5,0:5]
```

```python
# create the function for ordinal features dissimilarity 

def ordinal_disim(num1):
    max_rank = max(num1)
    min_rank = min(num1)
    disimilarities = []
    for x,y in itertools.combinations(num1, 2):
        disimilarities.append(abs(x-y)/(max_rank - min_rank))
    return disimilarities 
```

### Sets

Create a fuction that finds the similarities between sets 

```python
# transform the Groceries feature into a list of items and print the list

temp1 = list(groceries_data.Groceries)
items_list = [s.split(',') for s in temp1]
items_list
```

```python
# create a dictionary to be used in our function. For presentation purpose we print it horizontically row by row

mydict1 = dict(zip(cust_ids, items_list))
for x in mydict1:
    print (x,":",mydict1[x])
```

```python
# check the type of our dictionary 
type(mydict1)
```

```python
# create the jaccard similarity function

import itertools

#define jaccard similarity (size of intersection over size of union)
def jaccard_sim(list1, list2):
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(list1) + len(list2)) - intersection
    return float(intersection) / union
```

```python
# define the function for calculating the similarity matrix for set attributes
# create a vector for the dissimilarity values of our set attribute 'Groceries'

def set_disim(set1):
    disimilarities = []
    for pair in itertools.combinations(set1, 2):
        s=jaccard_sim(set1[pair[0]],set1[pair[1]])
        disimilarities.append(s)
    return disimilarities 

dis_df1_groceries = [1 - x for x in set_disim(mydict1)]
```

```python
# create square matrix for the set feature

from scipy.spatial.distance import *

disim_matrix_groceries = squareform(dis_df1_groceries)
disim_matrix_groceries
```

```python
# transform the square matrix into lower triangle matrix 

import numpy as np
df1_groceries = pd.DataFrame(np.tril(disim_matrix_groceries), columns=cust_ids, index=cust_ids) 
```

```python
# print the shape and a part of the matrix to check the rightness of the code

print(df1_groceries.shape)
df1_groceries.iloc[0:5,0:5]
```
